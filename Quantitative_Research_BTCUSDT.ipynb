```python
df = pd.read_csv(data)
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# Filter for January 2021
start_date = '2021-01-01'
end_date = '2022-01-31'
filtered_data = df[start_date:end_date]

# Display the filtered data
print(filtered_data.head())

```

```python
def bollinger_bands(df, window_size=29, num_of_std=2):
    rolling_mean = filtered_data['Close'].rolling(window=window_size).mean()
    rolling_std = filtered_data['Close'].rolling(window=window_size).std()

    filtered_data['Bollinger High'] = rolling_mean + (rolling_std * num_of_std)
    filtered_data['Bollinger Low'] = rolling_mean - (rolling_std * num_of_std)
    filtered_data['Rolling Median'] = filtered_data['Close'].rolling(window=window_size).median()
    return df
```

```python
filtered_data_with_bands = bollinger_bands(filtered_data)

plt.figure(figsize=(14,9))
plt.plot(filtered_data_with_bands['Close'], label='Close Price', alpha = 0.5)
plt.plot(filtered_data_with_bands['Bollinger High'], label='Bollinger High', color='red')
plt.plot(filtered_data_with_bands['Bollinger Low'], label='Bollinger Low', color='green')
plt.plot(filtered_data_with_bands['Rolling Median'], label='Rolling Median', color='yellow')
plt.title('Bollinger Bands for BTC/USDT')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
def calculate_profit_per_trade_with_percentages(df, initial_capital):
    capital = initial_capital
    capital_history = [capital]
    trade_profits = []
    trade_profit_percentages = []
    profitable_trades_count = 0

    for i in range(len(df)):
        if df['Position'].iloc[i] == 'Buy':
            buy_price = df['Close'].iloc[i]
            shares_bought = capital / buy_price
            capital = 0  # All capital invested

        elif df['Position'].iloc[i] == 'Sell' and buy_price is not None:
            sell_price = df['Close'].iloc[i]
            capital = shares_bought * sell_price
            capital_history.append(capital)
            trade_profit = capital - initial_capital
            trade_profit_percentage = (trade_profit / initial_capital) * 100

            trade_profits.append(trade_profit)
            trade_profit_percentages.append(trade_profit_percentage)

            if trade_profit > 0:
                profitable_trades_count += 1

            initial_capital = capital  # Update initial capital for next trade
            buy_price = None

    success_rate = (profitable_trades_count / len(trade_profits)) * 100 if trade_profits else 0

    return capital_history, trade_profits, trade_profit_percentages, success_rate

# Применяем функцию
initial_capital = 100000
capital_history, trade_profits, trade_profit_percentages, success_rate = calculate_profit_per_trade_with_percentages(filtered_data_with_bands, initial_capital)

# Распечатка результатов
for i, (profit, percentage) in enumerate(zip(trade_profits, trade_profit_percentages)):
    print(f"Trade {i+1}: Profit = {profit} ({percentage:.2f}%)")

print(f"Success rate of trading signals: {success_rate:.2f}%")

# Отображение истории капитала на графике
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(capital_history, marker='o', linestyle='-', color='b')
plt.title('Capital History Over Trades Starting from 100000')
plt.xlabel('Trade Number')
plt.ylabel('Capital')
plt.grid(True)
plt.show()
```

```python
import numpy as np

def calculate_unrealized_drawdown(df, initial_capital):
    capital = initial_capital
    peak_capital = initial_capital
    drawdowns = []
    max_unrealized_drawdown = 0
    buy_price = None
    shares_bought = 0

    for i in range(len(df)):
        if df['Position'].iloc[i] == 'Buy':
            buy_price = df['Close'].iloc[i]
            shares_bought = capital / buy_price
            capital = shares_bought * buy_price  # Capital is now tied up in shares

        elif df['Position'].iloc[i] == 'Sell' and buy_price is not None:
            sell_price = df['Close'].iloc[i]
            capital = shares_bought * sell_price
            buy_price = None  # Reset for the next trade
            shares_bought = 0

        # Update peak capital
        peak_capital = max(peak_capital, capital)

        # Calculate unrealized drawdown
        unrealized_drawdown = (peak_capital - capital) / peak_capital
        max_unrealized_drawdown = max(max_unrealized_drawdown, unrealized_drawdown)
        drawdowns.append(unrealized_drawdown)

    return drawdowns, max_unrealized_drawdown

# Apply the function
drawdowns, max_unrealized_drawdown = calculate_unrealized_drawdown(filtered_data_with_bands, initial_capital)

# Print results
print(f"Maximum Unrealized Drawdown: {max_unrealized_drawdown:.2f}")

```

```python
def calculate_bollinger_bands(data, n, w):
    rolling_mean = data['Close'].rolling(window=n).mean()
    rolling_std = data['Close'].rolling(window=n).std()
    upper_band = rolling_mean + (rolling_std * w)
    lower_band = rolling_mean - (rolling_std * w)
    return upper_band, lower_band
```

```python
def calculate_profit_with_capital(data, upper_band, lower_band, initial_capital):
    buy_price = None
    current_capital = initial_capital  # Использование отдельной переменной для текущего капитала
    total_profit = 0

    for i in range(len(data)):
        if buy_price is None and data['Close'].iloc[i] < lower_band.iloc[i]:
            buy_price = data['Close'].iloc[i]
            shares_bought = current_capital / buy_price
            current_capital = 0  # Весь капитал инвестирован
        elif buy_price is not None and data['Close'].iloc[i] > upper_band.iloc[i]:
            sell_price = data['Close'].iloc[i]
            current_capital = shares_bought * sell_price
            total_profit = current_capital - initial_capital  # Обновление общей прибыли
            buy_price = None

    return total_profit

best_profit = -np.inf
best_params = {'N': None, 'W': None}

for N in range(10, 31):
    for W in np.arange(1, 4, 0.5):
        upper_band, lower_band = calculate_bollinger_bands(filtered_data, N, W)
        profit = calculate_profit_with_capital(filtered_data, upper_band, lower_band, initial_capital)

        if profit > best_profit:
            best_profit = profit
            best_params = {'N': N, 'W': W}

print(f"Best Parameters: N={best_params['N']}, W={best_params['W']}, Profit={best_profit}")
```

```python
import numpy as np
import pandas as pd

# Functions for calculating Bollinger Bands and Profit
def calculate_bollinger_bands(data, n, w):
    rolling_mean = data['Close'].rolling(window=n).mean()
    rolling_std = data['Close'].rolling(window=n).std()
    upper_band = rolling_mean + (rolling_std * w)
    lower_band = rolling_mean - (rolling_std * w)
    return upper_band, lower_band

def calculate_profit(data, upper_band, lower_band):
    buy_price = None
    total_profit = 0

    for i in range(len(data)):
        if buy_price is None and data['Close'].iloc[i] < lower_band.iloc[i]:
            buy_price = data['Close'].iloc[i]
        elif buy_price is not None and data['Close'].iloc[i] > upper_band.iloc[i]:
            sell_price = data['Close'].iloc[i]
            total_profit += sell_price - buy_price
            buy_price = None

    return total_profit

best_total_profit = -np.inf
best_params = {'N_train': None, 'N_test': None, 'N': None, 'W': None}
initial_capital = 100000
cumulative_profits = [initial_capital]

N = len(df) - 1
N_train_options = [15, 30, 45, 60]
N_test_options = [10, 20, 30, 40]

# Итерация по периодам обучения и тестирования
for N_train in N_train_options:
    for N_test in N_test_options:
        # Рассчитываем размер шага для каждой комбинации
        step_size = N_train * 24

        # Общая прибыль для данной комбинации N_train и N_test
        total_profit_for_combination = 0

        # Итерация по набору данных с указанным размером шага
        for start in range(N, 0, -step_size):
            end_train = max(start - N_train * 24, 0)
            end_test = max(end_train - N_test * 24, 0)

            # Разбиение набора данных на обучающий и тестовый
            train_data = df.iloc[end_train:start]
            test_data = df.iloc[end_test:end_train]

            # Оптимизация параметров полос Боллинджера для обучающего набора
            for N in range(10, 31):
                for W in np.arange(1, 4, 0.5):
                    upper_band, lower_band = calculate_bollinger_bands(train_data, N, W)
                    profit = calculate_profit(test_data, upper_band, lower_band)

                    # Добавляем прибыль к общей сумме прибыли для данной комбинации
                    total_profit_for_combination += profit
                    cumulative_profits.append(profit)

        # Проверяем, является ли общая прибыль лучшей
        if total_profit_for_combination > best_total_profit:
            best_total_profit = total_profit_for_combination
            best_params = {'N_train': N_train, 'N_test': N_test, 'N': N, 'W': W}

cumulative_capital = np.cumsum(cumulative_profits)

# Вывод лучших параметров и прибыли
print(f"Best Parameters: N_train={best_params['N_train']}, N_test={best_params['N_test']}, Total Profit={best_total_profit}")

# Plotting the cumulative capital graph
plt.figure(figsize=(10, 6))
plt.plot(cumulative_capital)
plt.title("Cumulative Capital Over Time")
plt.xlabel("Trade Number")
plt.ylabel("Cumulative Profit")
plt.grid(True)
plt.show()

```
